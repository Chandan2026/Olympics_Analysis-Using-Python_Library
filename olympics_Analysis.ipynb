```python
goldMedals=df[(df.Medal=='Gold')]
goldMedals
```

```python
goldMedals=goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
goldMedals['ID'][goldMedals['Age']>50].count()
```

```python
goldMedals.region.value_counts().reset_index(name='Medal').head(10)
```

```python
totalGoldMedals=goldMedals.region.value_counts().reset_index(name='Medal').head(10)
Graph=sns.catplot(x='index',y='Medal',data=totalGoldMedals,height=5,kind='bar',palette='rocket')
Graph.despine(left=True)
Graph.set_xlabels('Top 5 countries')
Graph.set_ylabels('Number of Medals')
plt.title('Gold Medals per Country')
```
